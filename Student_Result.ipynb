```python
student_name = input("Enter student name: ")
math_marks = int(input("Enter maths marks: "))
science_marks = int(input("Enter science marks: "))
english_marks = int(input("Enter english marks: "))
Total_marks = math_marks+science_marks+english_marks
print(Total_marks)
if Total_marks >= 110:
    print("PASS")
elif Total_marks < 110:
    print("FAIL")
Average = Total_marks/3
print(Average)
if Average >= 75:
    print("A")
elif Average < 75 and Average >= 60:
    print("B")
elif Average < 60 and Average >=40:
    print("C")
elif Average < 0 and Average >=100:
    print("Invalid marks")

```
